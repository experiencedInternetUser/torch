### 1.1 Создание тензоров (7 баллов)
```python
# Создайте следующие тензоры:
# - Тензор размером 3x4, заполненный случайными числами от 0 до 1
# - Тензор размером 2x3x4, заполненный нулями
# - Тензор размером 5x5, заполненный единицами
# - Тензор размером 4x4 с числами от 0 до 15 (используйте reshape)
```

````python
tensor_3x4 = torch.rand((3, 4))

tensor_2x3x4 = torch.zeros((2, 3, 4))

tensor_5x5 = torch.ones((5, 5))

tensor_4x4 = torch.arange(0, 16, 1).reshape((4, 4))


print('Тензор размером 3x4, заполненный случайными числами от 0 до 1')
print(tensor_3x4)
print()
print('Тензор размером 2x3x4, заполненный нулями')
print(tensor_2x3x4)
print()
print('Тензор размером 5x5, заполненный единицами')
print(tensor_5x5)
print()
print('Тензор размером 4x4 с числами от 0 до 15')
print(tensor_4x4) 
````

### 1.2 Операции с тензорами (6 баллов)
```python
# Дано: тензор A размером 3x4 и тензор B размером 4x3
# Выполните:
# - Транспонирование тензора A
# - Матричное умножение A и B
# - Поэлементное умножение A и транспонированного B
# - Вычислите сумму всех элементов тензора A
```

````python
A = torch.rand((3, 4))
B = torch.rand((4, 3))

print('тензор A')
print(A)
print()
print('тензор B')
print(B)
print()

print('Транспонирование тензора A')
print(A.T)
print()
print('Матричное умножение A и B')
print(A @ B)
print()
print('Поэлементное умножение A и транспонированного B')
print(A * B.T)
print()
print('Вычислите сумму всех элементов тензора A')
print(A.sum())
````

### 1.3 Индексация и срезы (6 баллов)
```python
# Создайте тензор размером 5x5x5
# Извлеките:
# - Первую строку
# - Последний столбец
# - Подматрицу размером 2x2 из центра тензора
# - Все элементы с четными индексами
```

````python
tensor = torch.arange(0, 125, 1).reshape((5, 5, 5))

print('тензор 5x5x5')
print(tensor)
print()

print('Первая строка')
print(tensor[0, :, :])
print()
print('Последний столбец')
print(tensor[:, :, -1])
print()
print('Подматрица размером 2x2x2 из центра тензора')
print(tensor[2:4, 2:4, 2:4])
print()
print('Все элементы с четными индексами')
print(tensor[::2, ::2, ::2])
print()
````

### 1.4 Работа с формами (6 баллов)
```python
# Создайте тензор размером 24 элемента
# Преобразуйте его в формы:
# - 2x12
# - 3x8
# - 4x6
# - 2x3x4
# - 2x2x2x3
```

````python
tensor = torch.zeros(24)

print('2x12')
print(tensor.reshape((2, 12)))
print()
print('3x8')
print(tensor.reshape((3, 8)))
print()
print('4x6')
print(tensor.reshape((4, 6)))
print()
print('2x3x4')
print(tensor.reshape((2, 3, 4)))
print()
print('2x2x2x3')
print(tensor.reshape((2, 2, 2, 3)))
print()
````
